```python
import pandas as pd
import numpy as np

df = pd.read_csv("/kaggle/input/competitions/titanic/train.csv")

print(df.shape)
df.head()
```

```python
df['Survived'].value_counts()
```

```python
df.groupby(['Pclass', 'Sex'])['Survived'].mean()*100
```

Surival of women was the first priority on the titanic incident, even lower class of women had better chances than men  

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(data=df, x='Pclass', y='Survived', hue='Sex')
plt.title('Survival rate by class and Gender')
plt.ylabel('Survival rate')
plt.show()
```

```python
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title('Survival Rate by pessenger class')
plt.ylabel('Survival Rate')
plt.show()
```

This chart shows survival rate by passenger class,first class passengers had the highest survival rate at 63%, dropping to 24% for third class

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30)
plt.title('Age Destributions of Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.show()
```

Kids under 10 survived more often than older people, which means age also mattered—not just gender or class

```python
print("Average age of survivors:", round(df[df['Survived']==1]['Age'].mean(), 1))
print("Average age of non-survivors:", round(df[df['Survived']==0]['Age'].mean(), 1))
```
